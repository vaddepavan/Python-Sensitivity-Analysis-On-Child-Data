```python
"""
Title         :	Building a child language analyser
Author        :	VADDE PAVAN KUMAR
Student Id    :	29772990
Email Id      : pvad0001@student.monash.edu
Status        :	Final
Type          :	TASK - 1
Created       :	01-10-2018
last modified :	12-10-2018
Python-Version:	3.0

"""
import glob 
import re # importing the library files glob, regular expression
path_sli = 'ENNI Dataset/SLI/*.txt' # This is path where all dataset SLI files are there which are to be cleaned
files = glob.glob(path_sli) # this particular method return all the files present in the folder as a list
j = 1 # intializing the variables for creating the cleaned files 
k = 1
for filename in files: # loop for calling the list of files and collects each file and stores into filename each time
    with open(filename, 'r') as f: # this opens the file if it exist
        sli_list = [] # creating the list -- this list for storing the extracted lines
        new_sli_list = [] # this list for storing the cleaned lines of each file
        lines = f.readlines() # reads all the lines and stores in list format
        for i in range(0, len(lines)): # Proceesing through each line
            line = lines[i]
            if '*CHI:' in line: # if * CHI is found in the line and then the line will be saved to the list
                sli_list.append(line)
                ne = lines[i + 1] # you may want to check and then increment the line count
                if not ne.startswith('%mor:'): # to find whether the next line is also part of *CHI or not
                    sli_list.append(ne) # if it dosent starts with %mor and then it will append it to the list
        for line in sli_list: # this loop goes through each line in the list and then cleans the lines
            line = line.replace("<","") # replaces the particular chararcter with null and all the replace will happen
            line = line.replace(">","")
            line = line.replace("[+ bch]","")
            line = line.replace("[?]","")
            line = line.replace("[/-]","")
            line = line.replace("[!]","")
            line = line.replace("[x 3]","")
            line = line.replace("+/","")
            line = line.replace("[* m]","[*]")
            line = line.replace(" m:+ed","")
            line = line.replace("(.)","(.)")
            line = re.sub(r'\((\w+)\)',r'\1', line) # in this Regular expression checks for only the (any word) will be removed
            line = re.sub(r"\[: [a-z]+ ?[a-z]+ ?[a-z]+\]","", line) # this removes [: a b c]
            line = re.sub(r"\[\^ [a-zA-Z]+?[:] ?[a-z0-9]+ ?[a-z]+ ?[a-z]+\]","", line) # this removes [^ aorA: bor2 c d]
            line = re.sub(r"\[\^ [a-zA-Z]+ ?[a-z0-9]+ ?[a-z]+ ?[a-z]+\]","", line)# this removes [^ aorA bor2 c d]
            line = re.sub(r"\[\^ [a-z]+ ?[a-z]+ ?[a-z]+ \"(.+?)\"\]","", line) # this removes [^ a b c d]
            line = re.sub(r"\[\^ [a-z]+ ?[a-z]+ \"(.+?)\"\]","", line)# this removes [^ a b "anything with the format"] 
            line = re.sub(r"\[\^ [a-z]+ ?[a-z]+ ?[a-z]+ \'(.+?)\'\]","", line) # this removes [^ a b c 'anything with the format'] 
            line = re.sub(r"\[\^ [a-zA-Z]+\?\]","", line)# this removes [^ aorA?]
            line = re.sub(r"&=*[a-z]+","", line)# this removes &(=)*a, o or more occurence of o and any occurence a-z
            line = re.sub(r"\[\^ !\]","", line) # this removes [^ !]
            line = re.sub(r"\[\^ /[a-z]+/\]","", line)# this removes [^ /a/]
            new_sli_list.append(line) # after filtering each line in the file, each line will be append to the new sli list
        f = open("SLI_cleaned/SLI-%i.txt" %j,'w') # will be opening the file under the created folder as SLI-1.txt as write mode
        for item in new_sli_list: # travers through each item in the list
            f.write("%s" % item)# cleaned lines of a file stored to the new file created ex: SLI-1.txt
        f.close() # closing the opened file
        j = j+1 # counter for creating the file name automatically 
    f.close()  # closing the opened file
        


path_td = 'ENNI Dataset/TD/*.txt'
files = glob.glob(path_td)
for filename in files:
    with open(filename, 'r+') as f:
        TD_list = []
        new_TD_list = []
        lines = f.readlines()
        for i in range(0, len(lines)):
            line = lines[i]
            if '*CHI:' in line:
                TD_list.append(line)
                ne = lines[i + 1] # you may want to check that i < len(lines)
                if not ne.startswith('%mor:'):
                    TD_list.append(ne)
        for line in TD_list:
            line = line.replace("<","")
            line = line.replace(">","")
            line = line.replace("[+ bch]","")
            line = line.replace("[?]","")
            line = line.replace("[/-]","")
            line = line.replace("[!]","")
            line = line.replace("[x 3]","")
            line = line.replace("+..","")
            line = line.replace("+/","")
            line = line.replace("[* m]","[*]")
            line = line.replace(" m:+ed","")
            line = line.replace("(.)","(.)")
            line = re.sub(r'\((\w+)\)',r'\1', line)
            line = re.sub(r"\[: [a-z]+ ?[a-z]+ ?[a-z]+\]","", line)
            line = re.sub(r"\[\^ [a-zA-Z]+?[:] ?[a-z0-9]+ ?[a-z]+ ?[a-z]+ ?[a-zA-Z]+ ?[a-z]+ ?[a-z]+ ?[a-z]+\]","", line)
            line = re.sub(r"\[\^ ?[a-z]+ ?[a-z]+ ?[a-z]+ ?[a-z]+ ?[a-z]+ ?[a-z]+\]","", line)
            line = re.sub(r"\[\^ ?[a-zA-Z]+ ?[a-z]+\]","", line)
            line = re.sub(r"\[\^ [a-z]+ ?[a-z0-9]+ ?[a-z]+ ?[a-z]+\]","", line)
            line = re.sub(r"\[\^ [a-z]+ ?[a-z]+ ?[a-z]+ \"(.+?)\"\]","", line)
            line = re.sub(r"\[\^ [a-z]+ ?[a-z]+ \"(.+?)\"\]","", line)
            line = re.sub(r"\[\^ [a-z]+ ?[a-z]+ ?[a-z]+ \'(.+?)\'\]","", line)
            line = re.sub(r"\[\^ [a-zA-Z]+\?\]","", line)
            line = re.sub(r"&=*-*[a-z]+","", line)
            line = re.sub(r"\[\^ !\]","", line)
            line = re.sub(r"\[\^ \?\]","", line)
            line = re.sub(r"\[\^ /[a-z]+/\]","", line)
            new_TD_list.append(line)
        f = open("TD_cleaned/TD-%i.txt" %k,'w') 
        for item in new_TD_list:
            f.write("%s" % item)
        f.close()
        k = k+1
    f.close()
```
