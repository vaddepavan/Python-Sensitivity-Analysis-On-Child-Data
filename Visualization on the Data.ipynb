```python
"""
Title         :	Building a child language analyser
Author        :	VADDE PAVAN KUMAR
Student Id    :	29772990
Email Id      : pvad0001@student.monash.edu
Status        :	Final
Type          :	TASK - 1
Created       :	05-10-2018
last modified :	12-10-2018
Python-Version:	3.0

"""
from ast import literal_eval # importing the libraries 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re 

class Visualiser: # creating the visulaiser class 
    def __init__(self, data): # creating the constructor with the argument data 
        self.sli_list = [] # creating the list for storing the data 
        self.td_list = []
        self.sli_avg_dict = {} # creatign the dictionaries for storing the list for analysis 
        self.td_avg_dict = {}
        for i in range(0,10):
            line = data[0][i] # this collects the values from first list 
            line= literal_eval(line) # coolecting the data from the list which intern been string and being converted back to dictionary
            self.sli_list.append(line) # dictioanry is appended back to the list created
        for i in range(0,10):
            line = data[1][i]
            line= literal_eval(line)
            self.td_list.append(line)
        df1 = pd.DataFrame(self.sli_list, columns = ['length', 'unique', '[/]', '[//]','[*]','(.)']) # craeting the data frame for the list of dictionaries craeted 
        df2 = pd.DataFrame(self.td_list, columns = ['length', 'unique', '[/]', '[//]','[*]','(.)'])
        print("SLI - Statistics - average")
        display(df1) # dispalying the result avg of SLI stats in tabular format
        print("TD - Statistics - average")
        display(df2)  # dispalying the result avg of SLI stats in tabular format
    def compute_averages(self): # method is created for computing the averages of the stats 
        lists = [self.sli_list,self.td_list] # storing both the lsits in to one for caluclating 
        count = 0 
        for a in lists: # traversing through each list 
            total = 0
            for each in a: # travesring through each dictioanry 
                value = each.get('length') # matching with the key and the getting the value 
                total = total + value # clauclating the total of length of stats 
            avg_length = total // 10 # caluclating the average of statistics
            if count == 0: # counter variable for storing the values to the particualr dictionary 
                self.sli_avg_dict["avg_length"] = avg_length # avg have been saved to dictioaney with key value pair
            else:
                self.td_avg_dict["avg_length"] = avg_length

            total = 0 
            for each in a:
                value = each.get('unique')
                total = total + value
            avg_unique = total // 10
            if count == 0:
                self.sli_avg_dict["avg_unique"] = avg_unique
            else:
                self.td_avg_dict["avg_unique"] = avg_unique

            total = 0 
            for each in a:
                value = each.get('[/]')
                total = total + value
            avg_symbol1 = total // 10
            if count == 0:
                self.sli_avg_dict["avg_[/]"] = avg_symbol1
            else:
                self.td_avg_dict["avg_[/]"] = avg_symbol1

            total = 0 
            for each in a:
                value = each.get('[//]')
                total = total + value
            avg_symbol2 = total // 10
            if count == 0:
                self.sli_avg_dict["avg_[//]"] = avg_symbol2
            else:
                self.td_avg_dict["avg_[//]"] = avg_symbol2

            total = 0 
            for each in a:
                value = each.get('[*]')
                total = total + value
            avg_error = total // 10
            if count == 0:
                self.sli_avg_dict["avg_[*]"] = avg_error
            else:
                self.td_avg_dict["avg_[*]"] = avg_error

            total = 0 
            for each in a:
                value = each.get('(.)')
                total = total + value
            avg_pause = total // 10
            if count == 0:
                self.sli_avg_dict["avg_(.)"] = avg_pause
            else:
                self.td_avg_dict["avg_(.)"] = avg_pause
            count += 1  # incrementing the counter so the value will be stored into td dict
    def visualise_statistics(self): # method is constructed to perform visualization 
        names = ['avg_len', 'avg_unique', 'avg_[/]', 'avg_[//]', 'avg_[*]', 'avg_(.)'] # list of names have been created 
        values = self.sli_avg_dict.values() #getting the values of SLI - avaerages from the dictionary
        valuess = self.td_avg_dict.values() #getting the values of TD - avaerages from the dictionary
        print("SLI - Statistics - average: BAR GRAPH")
        plt.bar(range(len(self.sli_avg_dict)),values,tick_label=names) # plotting the bar graph for each SLI 
        plt.legend('SLI')
        plt.show() # shows the graph 
        print("TD - Statistics - average: BAR GRAPH")
        plt.bar(range(len(self.td_avg_dict)),valuess,tick_label=names) # plotting the bar graph for each TD
        plt.legend('TD')
        plt.show()
        #to visualize the both the graphs in one to compare 
        N = 6 # specify the no of stats to be dispalyed 
        ind = np.arange(N)  # numpy is used to arragne on the number 
        width = 0.27  #specifying the width of each bar graph 
        fig = plt.figure() # plot has been saved to the figure 
        ax = fig.add_subplot(111) # making it means the subplot 

        rects1 = ax.bar(ind, values, width, color='r') # calling as bar graph with position , width the bar and color 
        rects2 = ax.bar(ind+width, valuess, width, color='g')
        
        ax.set_ylabel('Average') # setting the labels  
        ax.set_xlabel('statistics')
        ax.set_xticks(ind+width) # these are used to specify the names on the y-axis for identifying for which is the bar graph
        ax.set_xticklabels(names) # on the x-axis
        ax.legend( (rects1[0], rects2[0]), ('SLI', 'TD') ) # craeting the legend for which we have plotted based on color
                  
        plt.show() # visualising using matplot lib

        
new_list = []      
f = open('SLI-stats.txt','r') # reading the values of stats from the stored file 
lines = f.readlines()
new_list.append(lines) # appending it to the list 
f1 = open("TD-stats.txt")
line = f1.readlines()
new_list.append(line) # appending it to the list 
 
visualiser = Visualiser(new_list) #creating the visulaiser and passing the arument as the list 
visualiser.compute_averages() # calling the averages method
visualiser.visualise_statistics() # calling the visualise method
```
